```
from pyspark import SparkConf, SparkContext
from pyspark.sql.functions import *
```

```
path = "/FileStore/tables/fakefriends.csv"
```

```
df = spark.read.csv(path)

display(df, header="False", inferSchema="True")
```

```
display(df.describe())
```

```
display(df.printSchema())
```

```
df = df.withColumn("No_of_friends", col('_c3').cast("integer"))
df = df.withColumn("Age", col('_c2').cast("integer"))
```

```
cols_to_drop = ['_c2', '_c3','_c0']

df = df.drop(*cols_to_drop)

display(df)
```

```
grouped_df = df.groupBy(col('Age')).agg(avg('No_of_friends'))

display(grouped_df)
```

```
display(grouped_df.where(grouped_df.Age == 55))
```
